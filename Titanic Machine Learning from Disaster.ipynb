The data has been already split into training and test data. Our first step is to clean the data and Since it is extremely important to do the same changes in both data set we will start by defining a function to serve this purpose.

<p> We note that the PassengerId and the ticket number are not important, since what matters regarding those features is the class of the passengers ended up in. Furthermore, the cabin location depends on the class, (1st class passengers are located to same part of the boat), thus we will also remove the cabin column. Finaly we are not interested in the Port of Embarkation. Due to the hight corrolation will will keep the Fare column. The Fare price indicates the class but, people in the same calss that had pricier tickets maybe ended up with more favorably located cabins. Therefore we will not remove this column.

```python
def data_changes(data):
    
    data.drop(labels=['PassengerId','Ticket','Cabin','Embarked','Name'],axis=1,inplace=True)
    
    data["Age"] = data["Age"].fillna(round(data['Age'].mean()))
    data["Fare"] = data["Fare"].fillna(round(data['Fare'].mean()))
    
    
    data.replace(('male','female'),(0,1), inplace=True)
    #data.replace(('C','Q','S'),(0,1,2), inplace=True)
    
    return data
```

## We will try different ML methods

```python
X_train = train.drop("Survived",axis=1)
y_train = train["Survived"]
X_test  = test
```

### SVC

```python
from sklearn.svm import SVC
from sklearn import cross_validation as cval
from sklearn import svm 
```

```python
model=svm.SVC(kernel='linear',C=0.1,gamma=0.1)
model.fit(X_train,y_train)
prediction_svc=model.predict(test)
```

```python
print(model.score(X_train, y_train))
```

### Decision Tree

```python
from sklearn import tree
```

```python
model1=tree.DecisionTreeClassifier()
model1.fit(X_train, y_train)
pred_decTree=model1.predict(test)
print(model1.score(X_train, y_train))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(X_train, y_train)

pred_rf = random_forest.predict(test)

cv_random_forest=  cval.cross_val_score(random_forest, X_train, y_train, cv=10) 
```

```python
print(random_forest.score(X_train, y_train))
print(cv_random_forest.mean())
```

### K-Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
pred_knc = knn.predict(X_test)
print(knn.score(X_train, y_train))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2=LogisticRegression()
model2.fit(X_train, y_train)
pred_logreg = model2.predict(X_test)

print(model2.score(X_train, y_train))
```

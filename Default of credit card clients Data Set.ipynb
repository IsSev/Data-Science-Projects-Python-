##  Default of credit card clients Data Set 


<p> The data are from <a href="http://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients">UCI</a>



### Attribute Information:

This dataset employed a binary variable, 

<b>"D_PAY" </b> default payment (Yes = 1, No = 0), as the response variable. 

In this dataset we use the following 23 variables as explanatory variables: 

<b>LIMIT_BAL </b> Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit. 

<b>SEX</b> Gender (1 = male; 2 = female). 

<b>EDUCATION </b> Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

<b>MARRIAGE</b> Marital status (1 = married; 2 = single; 3 = others). 

<b>AGE:</b> Age (year). 

<b>PAY_1 - PAY_6</b> History of past payment. 
We tracked the past monthly payment records (from April to September, 2005) as follows: 
<ul>
  <li>PAY_1 = the repayment status in September, 2005</li>
  <li>PAY_2 = the repayment status in August, 2005</li>
  <li>............................................</li>
  <li>PAY_6 = the repayment status in April, 2005</li>
</ul>  
The measurement scale for the repayment status is:
<ul>
  <li>-2 = 0 bill </li>
  <li>-1 = pay duly</li>
  <li>0 = partial pay</li>
  <li>1 = payment delay for one month</li>
  <li>2 = payment delay for two months</li>
  <li>................................</li>
  <li>8 = payment delay for eight months</li>
  <li>9 = payment delay for nine months and above</li>
</ul>


<b>BILL_AMT1-BILL_AMT6</b> Amount of bill statement in NT dollars
<ul>
  <li>PBILL_AMT1 = amount of bill statement in September, 2005</li>
  <li>BILL_AMT2 = amount of bill statement in August, 2005</li>
  <li>...................................................</li>
  <li>BILL_AMT6= amount of bill statement in April, 2005</li>
</ul>

<b>PAY_AMT1-PAY_AMT6</b> Amount of previous payment in NT dollars
<ul>
  <li>PAY_AMT1 = amount paid in September, 2005</li>
  <li>PAY_AMT2 = amount paid in August, 2005</li>
  <li>......................................</li>
  <li>PAY_AMT6 = amount paid in April, 2005</li>
</ul>

## We will try different ML methods to predict default payment

Firstly, we have to split the data in training and testing

```python
y = df['D_PAY'].copy()
X=df.drop(labels=['D_PAY'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
model=LogisticRegression()
model.fit(x_train, y_train)
pred=model.predict(x_test)

print('The Logistic Regression training score is: ',model.score(x_train,y_train)) 
print('The Logistic Regression testing score is: ',model.score(x_test,y_test))
print("The accuracy of the model is: ",accuracy_score(y_test,pred, normalize=False))
```

```python
from mlxtend.plotting import plot_confusion_matrix
confusion=metrics.confusion_matrix(y_test, pred)
fig, ax = plot_confusion_matrix(conf_mat=confusion)
plt.show()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100)

random_forest.fit(x_train, y_train)
pred_rf = random_forest.predict(x_test)
print('The random forest training score is: ',random_forest.score(x_train,y_train)) 
print('The random forest testing score is: ',random_forest.score(x_test,y_test)) 
print("the accuracy of the model is: ",accuracy_score(y_test,pred_rf, normalize=False))
```

```python
confusion=metrics.confusion_matrix(y_test, pred_rf)
fig, ax = plot_confusion_matrix(conf_mat=confusion)
plt.show()
```

### KNeighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
pred_knn = random_forest.predict(x_test)
print('The KNeighbors training score is: ',knn.score(x_train,y_train)) 
print('The KNeighbors testing score is: ',knn.score(x_test,y_test)) 
print("the accuracy of the model is: ",accuracy_score(y_test,pred_knn, normalize=False))
```

```python
confusion=metrics.confusion_matrix(y_test, pred_knn)
fig, ax = plot_confusion_matrix(conf_mat=confusion)
plt.show()
```

### Neural Network

The neural network may have difficulty converging before the maximum number of iterations allowed if the data is not normalized. Thus we will use the built-in StandardScaler for standardization.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(24,24,24))
```

```python
mlp.fit(x_train,y_train)
```

```python
predictions = mlp.predict(x_test)
print("the accuracy of the model is: ",accuracy_score(y_test,predictions, normalize=False))
```

```python
confusion=metrics.confusion_matrix(y_test, predictions)
fig, ax = plot_confusion_matrix(conf_mat=confusion)
plt.show()
```
